```python
import numpy as np
import pandas as pd

import json
```

```python
with open("models/results.json", "r") as f:
    results = json.load(f)
```

```python
table = []
for result in results:
    rf, xgb = {}, {}
    rf["id"] = result["id"] + "_rf"
    xgb["id"] = result["id"] + "_xgb"
    rf_predict = (np.array(result["y_proba_rf"]) >= 0.5).astype(int)
    xgb_predict = (np.array(result["y_proba_xgb"]) >= 0.5).astype(int)
    rf["accuracy"] = np.mean(np.array(result["y_test"]) == rf_predict)
    xgb["accuracy"] = np.mean(np.array(result["y_test"]) == xgb_predict)
    table.append(rf)
    table.append(xgb)
```

```python
df = pd.DataFrame(table)
df.sort_values("accuracy", ascending=False)
```
