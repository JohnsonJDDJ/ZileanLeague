```python
from zilean import SnapShots
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
snaps = SnapShots("data/match_8_15.csv")
data = pd.DataFrame(snaps.summary())
data = data.drop(["matchId"], axis=1)
data["win"] = data["win"].astype(int)
train, test= train_test_split(data, test_size=0.2)
```

```python
scaler = StandardScaler()
columns = train.columns
train = scaler.fit_transform(train)
train = pd.DataFrame(train, columns=columns)
```

```python
train["temp"] = train["totalGold_1_frame15"] - train["totalGold_1_frame8"]
sns.histplot(x="temp", data=train, hue="win")
```
